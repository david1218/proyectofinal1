```python
import pandas as pd
from bokeh.core.properties import value
from bokeh.io import show, output_file, output_notebook
from bokeh.models import ColumnDataSource
#from bokeh.palettes import Spectral6, viridis
from bokeh.palettes import *
from bokeh.transform import factor_cmap
from bokeh.plotting import *
from bokeh.plotting import figure
output_notebook()
data=pd.read_csv("Libro1.csv",sep=";", encoding = 'ISO 8859-1')  
data
```

```python
Jugadores=data["Nombres de jugadores"].tolist()
Total=data["Total de Goles"].tolist()
#Porte=[1.80, 1.75, 1.80, 1.79, 1.69, 1.74, 1.86, 1.85, 1.86, 1.74, 1.80, 1.70, 1.67, 1.86, 1.64, 1.90, 1.85, 1.75, 1.80, 1.80 ]
Colores=viridis(len(Jugadores))
print(Colores)
```

```python
p = figure(x_range=Jugadores, plot_height=500, plot_width=1100, title="JUGADORES/TOTAL GOLES",
           toolbar_location=None, tools="")

p.vbar(x=Jugadores, top=Total, width=0.9, color=Colores)

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.major_label_orientation = 1.4
show(p)
```

```python
Jugadores=data["Nombres de jugadores"].tolist()
Porte=data["Estatura"].tolist()
```

```python
# sorting the bars means sorting the range factors
sorted_Jugadores = sorted(Jugadores, key=lambda x: Porte[Jugadores.index(x)])

p = figure(x_range=sorted_Jugadores, plot_height=400, plot_width=1100, title="JUGADORES ESTATURAS",
           toolbar_location=None, tools="")

p.vbar(x=Jugadores, top=Porte, width=0.7, color=Colores)
p.xaxis.major_label_orientation = 1.4
p.xgrid.grid_line_color = None
p.y_range.start = 0
show(p)
```

```python
Jugadores=data["Nombres de jugadores"].tolist()
Edad=data["Edad"].tolist()
Colores = viridis(len(Jugadores))
```

```python
source = ColumnDataSource(data=dict(Jugadores=Jugadores, Edad=Edad))

p = figure(x_range=Jugadores, plot_height=500, plot_width=1100, toolbar_location=None, title="EDADES DE JUGADORES")
p.vbar(x='Jugadores', top='Edad', width=0.5, source=source, legend="Jugadores",
       line_color='white', fill_color=factor_cmap('Jugadores', palette=Colores, factors=Jugadores))
        
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 40
p.legend.orientation = "vertical"
p.legend.location = "top_right"
p.xaxis.major_label_orientation = 1.4
show(p)
```

```python
from bokeh.transform import dodge
from bokeh.core.properties import value
Nombres = ['Lionel Messi', 'Luis Suárez', 'Jordi Alba', 'Rafinha']
Temporadas = ['2015', '2016']

data = {'Nombres' : Nombres,
        '2015'   : [15, 0, 2, 7],
        '2016'   : [9, 2, 0, 0]}

source = ColumnDataSource(data=data)

p = figure(x_range=Nombres, y_range=(0, 9), plot_height=400, plot_width=1000, title="MEJORES JUGADORES DE LA TEMPORADA GOLES",
           toolbar_location=None, tools="")

p.vbar(x=dodge('Nombres', -0.25, range=p.x_range), top='2015', width=0.6, source=source,
       color="#c9d9d3", legend=value("Temporada1"))

p.vbar(x=dodge('Nombres',  0.0,  range=p.x_range), top='2016', width=0.6, source=source,
       color="#718dbf", legend=value("Temporada2"))


p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 18
p.legend.orientation = "horizontal"
p.legend.location = "top_right"
p.xaxis.major_label_orientation = 0
show(p)
```
